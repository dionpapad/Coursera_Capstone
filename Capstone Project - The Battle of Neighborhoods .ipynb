## At this point it is obvious that the results need some shorting. I am keeping only the venue name, the category and the geo coordinates, since that's what I am interested for.

```python
from pandas.io.json import json_normalize
venues=results['response']['groups'][0]['items']
venuesNearBy=json_normalize(venues)
filtered_columns=['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
venuesNearBy=venuesNearBy.loc[:, filtered_columns]
venuesNearBy.columns = [col.split(".")[-1] for col in venuesNearBy.columns]
venuesNearBy
```

## After shorting data in dataframe, I am presenting them inside a map. The map is created by using the geo coordinations extracted from the csv file, while foursquare API gave the geoo coord for the venues. Also, the venues can be seen in blue dots.

```python
import folium
tlat=city_latitude
tlong-city_longitude
cityMap=folium.Map(location=[tlat,tlong],zoom_start=15)

for lat,long,category,name in zip(venuesNearBy['lat'],venuesNearBy['lng'],venuesNearBy['categories'],venuesNearBy['name']):
    label='{}, {}'.format(name, category)
    label=folium.Popup(label,parse_html=True)
    folium.CircleMarker([lat,long],
                       radius=5,
                       popup=label,
                       color='blue',
                       fill=True,
                       fill_color='#3186cc',
                       fill_opacity=0.7,
                       parse_html=False).add_to(cityMap)
    
cityMap
```
